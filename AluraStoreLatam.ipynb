### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Agregar columna para identificar cada tienda
tienda1["tienda"] = "Tienda 1"
tienda2["tienda"] = "Tienda 2"
tienda3["tienda"] = "Tienda 3"
tienda4["tienda"] = "Tienda 4"

# Concatenar en un solo dataframe
tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

tiendas.head()
```

```python
# Facturación total por tienda (Precio * Cantidad de cuotas)
tiendas["facturacion"] = tiendas["Precio"] * tiendas["Cantidad de cuotas"]

facturacion_tiendas = tiendas.groupby("tienda")["facturacion"].sum().reset_index()

print("Facturación total por tienda:")
print(facturacion_tiendas)

# Tienda con menor facturación
tienda_menos = facturacion_tiendas.loc[facturacion_tiendas["facturacion"].idxmin()]
print("\nLa tienda con MENOR facturación es:", tienda_menos["tienda"],
      "con", tienda_menos["facturacion"])
```

# 2. Ventas por categoría

```python
ventas_categoria = tiendas.groupby("Categoría del Producto")["Cantidad de cuotas"].sum().reset_index()

print("Ventas por categoría:")
print(ventas_categoria.sort_values(by="Cantidad de cuotas", ascending=False))
```

# 3. Calificación promedio de la tienda


```python
calificacion_prom = tiendas.groupby("tienda")["Calificación"].mean().reset_index()

print("Calificación promedio por tienda:")
print(calificacion_prom)
```

# 4. Productos más y menos vendidos

```python
# Más vendidos
mas_vendidos = tiendas.groupby("Producto")["Cantidad de cuotas"].sum().sort_values(ascending=False).head(10)

# Menos vendidos
menos_vendidos = tiendas.groupby("Producto")["Cantidad de cuotas"].sum().sort_values().head(10)

print("Top 10 productos más vendidos:\n", mas_vendidos)
print("\nTop 10 productos menos vendidos:\n", menos_vendidos)
```

# 5. Envío promedio por tienda

```python
envio_promedio = tiendas.groupby("tienda")["Costo de envío"].mean().reset_index()

print("Envío promedio por tienda:")
print(envio_promedio)
```
